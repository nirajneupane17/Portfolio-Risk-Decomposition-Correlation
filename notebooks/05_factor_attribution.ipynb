```python
import sys; sys.path.insert(0,"../src")
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from factor_model import fit_factor_model, portfolio_factor_attribution
rets=pd.read_csv("../data/returns.csv",index_col="Date",parse_dates=True)
factors=pd.read_csv("../data/factor_data.csv",index_col="Date",parse_dates=True)
assets=list(rets.columns); print("OK")
```

```python
[print(f"{a}: R²={fit_factor_model(rets[a],factors)['r2']:.3f}  Sys={fit_factor_model(rets[a],factors)['systematic_pct']:.0f}%") for a in assets]
```

```python
w=pd.read_csv("../data/portfolio_weights.csv",index_col="Asset")["equal_weight"].values
port=(rets*w).sum(axis=1); pa=portfolio_factor_attribution(port,factors)
print(f"Port R²: {pa['r2']:.3f}  Systematic: {pa['systematic_pct']:.1f}%  Idio: {pa['idiosyncratic_pct']:.1f}%")
print("\nFactor Vol Contrib:"); [print(f"  {f}: {v:.4f}%") for f,v in pa["factor_vol_contrib"].items()]
```
