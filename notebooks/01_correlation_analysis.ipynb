```python
import sys; sys.path.insert(0,"../src")
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from correlation_engine import rolling_correlation, regime_correlation, pca_concentration
print("OK")
```

```python
rets=pd.read_csv("../data/returns.csv",index_col="Date",parse_dates=True)
print(f"Shape: {rets.shape}")
```

```python
mc=rolling_correlation(rets)
print(f"Mean corr: {mc.mean():.3f} | COVID peak: {mc.loc["2020-02-01":"2020-05-31"].max():.3f}")
```

```python
periods={"Full":slice(None),"COVID":slice("2020-02-01","2020-05-31"),"Calm2023":slice("2023-01-01","2023-12-31")}
for nm,c in regime_correlation(rets,periods).items():
    u=c.values[np.triu_indices_from(c.values,k=1)]
    print(f"{nm}: avg={u.mean():.3f}")
```
